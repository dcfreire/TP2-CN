```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("tests/its.json") as fp:
    data = json.load(fp)
```

```python
plt.figure(figsize=(15,8))


for experiment in data:
    plt.plot(np.mean(np.array(experiment["results"]), axis=0), label=experiment["its"])

plt.xlabel("Iteration")
plt.ylabel("Best Solution")
plt.legend(title="Max Its")
plt.show()

```

```python
plt.figure(figsize=(15,8))

heights = []
bars = []
for experiment in data:
    heights.append(experiment["time"])
    bars.append(experiment["its"])

plt.xlabel("Iterations")
plt.ylabel("Time (s)")
plt.plot(bars, heights)
plt.legend()
plt.show()
```

```python
import pandas as pd
```

```python
def get_best(data):
    b = []
    for experiment in data:
        for result in experiment["results"]:
            b.append(result[-1][0])
    return b
```

```python
with open("tests/final/2.json") as fp:
    data = json.load(fp)

best = get_best(data)
df = pd.DataFrame(best).describe().transpose()
df
```
